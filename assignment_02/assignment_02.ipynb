# 1. Prompts a user to enter 10 integers.


```python
the_list = []

for i in range(0,10):
    while True:
        x = input("Please enter an integer:")
        try:
            y = int(x)
            the_list.append(y)
            print("got integer:", y)
            break
        except:
            print("You must enter an integer. Please try again.")


print("You have entered:", the_list)
```

```python
max = the_list[0]
for x in the_list:
  if x > max:
    max = x
print(max)
```

```python
min  = the_list[0]
for x in the_list:
  if x < min:
    min = x
print(min)
```

```python
Range= max-min
print("Range is:", Range)
```

```python
total = 0

for i in the_list:
    total += i    # total = total + i

print("total = ", total)
```

```python
mean = total / len(the_list)
print(mean)
```

```python
 deviations = [(x - mean) ** 2 for x in the_list]
 n=len(the_list)
variance = sum(deviations) / n
print(variance)
```

```python
SD= variance ** 0.5
print("Standard Deviation is:",SD)
```
